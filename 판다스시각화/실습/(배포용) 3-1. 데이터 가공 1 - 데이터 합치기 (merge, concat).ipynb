# 4. 데이터 가공 1 - 데이터 통합
## 4-1. pandas.merge()

```python
#샘플 데이터 생성
df1 = DataFrame([['홍길동',23,'남'],['이순신',55,'남'], ['신사임당',48,'여']], 
                           columns = ['이름','나이','성별'])
df2 = DataFrame([['홍길동','터미네이터',5],['홍길동','비긴어게인',2], 
                 ['홍길동','살인의추억',4], ['신사임당','비긴어게인',5],
                ['장보고','살인의추억',3]], 
                           columns = ['이름','영화','평점'])
```

```python
pd.merge(df1,df2) # df1, df2에서 장보고, 이순신은 없어서 안뜸!
# inner 방식: 두개 모두 있는 경우에만 출력
```

```python
pd.merge(df1,df2,how='outer')
# outer 방식: 둘 중 하나에만 있어도 출력
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

## 4-2. pandas.concat(axis = 0)
* 동일한 컬럼들을 가지고 있는 데이터프레임들을 하나로 합치고 싶은 경우
- pd.concat([df1, df2] , axis = 0 )
- 첫번째 인자로 object: 내가 합치고자 하는 데이터들의 목록을 리스트의 형태로 나열

```python
df1 = DataFrame([['홍길동',23,'남'],['이순신',55,'남']], 
                           columns = ['이름','나이','성별'])
df2 = DataFrame([['신사임당',48,'여'], ['장보고',22,'남']], 
                           columns = ['이름','나이','성별'])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2],axis = 0 ) # row index 가 본래 값을 가져온다
```

```python
pd.concat([df1,df2],axis = 0, ignore_index = True) #기존의 index 무시
```

## 4-3. concat(axis = 1)
* 서로 다른 컬럼을 가진 데이터프레임들을 로우 인덱스의 값을 기준으로 합치고 싶은 경우

```python
# 샘플 데이터
df1 = DataFrame([[23,'남'],[55,'남'], [48,'여']], 
                           columns = ['나이','성별'], index = ['홍길동','이순신','신사임당'])
df2 = DataFrame([['서울',3], ['경기', 1]], 
                           columns = ['주소','학년'], index = ['홍길동','신사임당'])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis = 1)
```

```python
pd.concat([df1,df2],axis = 1,join = 'inner')
```

## 실제 데이터 실습
* data/NC Dinos.xlsx 파일에서 2013, 2014, 2015시즌 정보를 읽어서, 
* 각각 NC13, NC14, NC15에 저장한 후, 이를 NCAll이라는 하나의 데이터프레임으로 합치기


```python
#sheet_name = None 인 경우, 모든 sheet 각각의 데이터프레임의 형태로 가져옴
NC = pd.read_excel("data/NC DInos.xlsx",sheet_name = None) 
NC
```

```python
NC.keys()
```

```python
NC.values()
```

```python
#각각 NC13, NC14, NC15에 저장
NC13 = NC['2013']
NC14 = NC['2014']
NC15 = NC['2015']
```

```python
NC13['년도'] = 2013
NC14['년도'] = 2014
NC15['년도'] = 2015
```

```python
#NCALL라는 데이터프레임으로 합치기
NCALL = pd.concat([NC13,NC14,NC15],axis = 0, ignore_index= True)
NCALL
```
