## 4-1. pandas.merge() 
 데이터베이스의 join 연산과 유사함

```python
#샘플 데이터 생성
df1 = DataFrame([['홍길동',23,'남'],['이순신',55,'남'], ['신사임당',48,'여']], 
                           columns = ['이름','나이','성별'])
df2 = DataFrame([['홍길동','터미네이터',5],['홍길동','비긴어게인',2], 
                 ['홍길동','살인의추억',4], ['신사임당','비긴어게인',5],
                ['장보고','살인의추억',3]], 
                           columns = ['이름','영화','평점'])
```

```python
pd.merge(df1,df2) #df1.merge(df2) #장보고와 이순신은 사용이 X 
```

```python
pd.merge(df1,df2, how = 'outer') #df1.merge(df2, how = 'outer')
```

```python
pd.merge(df1,df2, how = 'left') #df1.merge(df2, how = 'left')
```

```python
pd.merge(df1,df2, how = 'right',on = "이름") # #df1.merge(df2, how = 'right')
```

### 오류를 확인하자!

#### 오류1 ) 테이블 두 곳 중 한 곳에만 있는 컬럼을 사용 시에는 오류가 발생

```python
pd.merge(df1,df2, how = 'right',on = "나이") 
```

#### 오류2)컬럼명이 다른 경우 컬럼명을 left_on, right_on을 이용하여 설정 필요. => 이후 한 컬럼은 drop!

```python
#샘플 데이터 생성
df1 = DataFrame([['홍길동',23,'남'],['이순신',55,'남'], ['신사임당',48,'여']], 
                           columns = ['이름','나이','성별'])
df2 = DataFrame([['홍길동','터미네이터',5],['홍길동','비긴어게인',2], 
                 ['홍길동','살인의추억',4], ['신사임당','비긴어게인',5],
                ['장보고','살인의추억',3]], 
                           columns = ['이름1','영화','평점'])
```

```python
pd.merge(df1,df2) #오류
```

```python
pd.merge(df1,df2,left_on='이름',right_on="이름1").drop("이름1",axis = 1) #이름1은 drop 해주면 된다 :)
```

#### 오류4) 공통된 컬럼이 두 개 이상인 경우에는 컬럼 하나만 선택하여 코드 실행

```python
#샘플 데이터 생성
df1 = DataFrame([['홍길동',23,'남'],['이순신',55,'남'], ['신사임당',48,'여']], 
                           columns = ['이름','평점','성별'])
df2 = DataFrame([['홍길동','터미네이터',5],['홍길동','비긴어게인',2], 
                 ['홍길동','살인의추억',4], ['신사임당','비긴어게인',5],
                ['장보고','살인의추억',3]], 
                           columns = ['이름','영화','평점'])
```

```python
pd.merge(df1,df2) #공통된 컬럼이 두 개 이상인 경우 이름과 평점을 둘 다 확인한다! => 하나 선택 가능
```

```python
pd.merge(df1,df2,on = "이름") 
```

## 4-2. pandas.concat(axis = 0)
* 동일한 컬럼들을 가지고 있는 데이터프레임들을 하나로 합치고 싶은 경우 사용 *concat 시에는 컬럼명을 통일하라!

```python
df1 = DataFrame([['홍길동',23,'남'],['이순신',55,'남']], 
                           columns = ['이름','나이','성별'])
df2 = DataFrame([['신사임당',48,'여'], ['장보고',22,'남']], 
                           columns = ['이름','나이','성별'])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2],axis = 0) #row index가 본래 값을 가져옴
```

```python
pd.concat([df1,df2],axis = 0, ignore_index = True) #기존의 index 무시
```

## 4-3. concat(axis = 1)
* 서로 다른 컬럼을 가진 데이터프레임들을 로우 인덱스의 값을 기준으로 합치고 싶은 경우

```python
# 샘플 데이터
df1 = DataFrame([[23,'남'],[55,'남'], [48,'여']], 
                           columns = ['나이','성별'], index = ['홍길동','이순신','신사임당'])
df2 = DataFrame([['서울',3], ['경기', 1]], 
                           columns = ['주소','학년'], index = ['홍길동','신사임당'])
```

```python
df1
```

```python
df2
```

* [df1,df2] / [df2,df1] 리스트 순서에 따라 return 값 상이

```python
pd.concat([df1,df2],axis = 1)
```

```python
pd.concat([df2,df1],axis = 1)
```

* 한 쪽에만 있는 데이터는 제외시키고 싶은 경우 (merge inner 같이) => join 사용

```python
pd.concat([df1,df2],axis = 1, join = 'inner')
```

```python
pd.concat([df1,df2],axis = 1, join = 'outer') #default
```

* 중복된 index를 가지는 경우 에러 발생 ex) df1 홍길동 이름 중복

```python
df1 = DataFrame([[23,'남'],[55,'남'], [48,'여']], 
                           columns = ['나이','성별'], index = ['홍길동','홍길동','신사임당'])
df2 = DataFrame([['서울',3], ['경기', 1]], 
                           columns = ['주소','학년'], index = ['홍길동','신사임당'])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2],axis=1)
```

## 실제 데이터 실습

* 각각 NC13, NC14, NC15에 저장한 후, 이를 NCAll이라는 하나의 데이터프레임으로 합치기


```python
#NC Dinos.xlsx 데이터 불러오기
NC = pd.read_excel("data/NC Dinos.xlsx", sheet_name=None)
```

```python
#데이터 확인
type(NC)
```

```python
NC.keys()
```

```python
NC.values()
```

```python
#각각 NC13, NC14, NC15에 저장
NC13 = NC['2013']
NC14 = NC['2014']
NC15 = NC['2015']
```

```python
#NCALL라는 데이터프레임으로 합치기
NCALL = pd.concat([NC13,NC14,NC15],axis = 0, ignore_index= True)
NCALL
```
